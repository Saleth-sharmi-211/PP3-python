```python
def letter(text):
 vowels={'a',"e","i","o","u"}
 count=0
 for i in text.lower():
  if i in vowels:
    count += 1
 return count
print (letter("sharmi"))

```

```python
students={
    "anu":[78,85,90],
    "bala":[60,72,68],
    "charan":[88,91,84],
    "divya":[45,50,40]
}
def calculate_average(marks):
  return sum(marks)/len(marks)
check_result=lambda avg:"pass"if avg>=50 else"fail"
student_result={}
for name,marks in students.items():
  avg=round(calculate_average(marks),2)
  result=check_result(avg)
  student_result[name]={
      "average":avg,
      "result":result
  }
  print("passed students:")
  for name,details in student_result.items():
    if details["result"]=="pass":
      print(f"{name}-average:{details['average']}")
  topper=max(student_result,key=lambda x:student_result[x]["average"])
  print("\ntopper:",topper)
```

```python
def count_digits(n):
 if n==0:
     return 0
 return 1+count_digits(n//10)
print(count_digits(424))

def sum_of_digits(n):
  if n==0:
    return 0
  return (n%10)+sum_of_digits(n//10)
print(sum_of_digits(354))

def max_digit_rec(n):
 if n<10:
  return n
 return max(n%10,max_digit_rec(n//10))
print(max_digit_rec(58))

def reverse(n,rev=0):
 if n==0:
    return rev
 return reverse(n//10,rev*10+n%10)
print(reverse(12345))
```
